# Functions

```python
# Function to show survivors by group of a column
def survival_rate(df_orig,column,t):
    df=pd.DataFrame()
    df['Total']=df_orig.groupby(column).size()
    df['Survived'] = df_orig.groupby(column).sum()['Survived']
    df['Percentage'] = np.round(df['Survived']/df['Total']*100,2)
    print(df)

    df['Survived'].plot(kind=t)
    df['Total'].plot(kind=t,alpha=0.5,title="Survivors per "+str(column))
    plt.show()
```

```python
def normalize_age_lass_than_one(age):
    if age < 1:
        return 1
    else:
        return age
```

```python
def groupby_age(age):
    if age <= 15:
        return "(1) 0-15"
    elif age <= 30:
        return "(2) 15-30"
    elif age <= 45:
        return "(3) 30-45"
    elif age <= 60:
        return "(4) 45-60"
    elif age <= 75:
        return "(5) 60-75"
    elif age <= 90:
        return "(6) 75-90"
    else:
        return "(7) NaNs"
```

```python
def groupby_fare(fare):
    if fare <= 25.0:
        return "(1) 0-25"
    elif fare <= 50.0:
        return "(2) 25-50"
    elif fare <= 75.0:
        return "(3) 50-75"
    elif fare <= 100.0:
        return "(4) 75-100"
    elif fare <= 200.0:
        return "(5) 100-200"
    elif fare <= 300.0:
        return "(6) 200-300"
    elif fare > 300.0:
        return "(7) 300-higher"
```

```python
def groupby_family_members(members):
    if members == 0:
        return "(1) 0"
    elif members == 1:
        return "(2) 1"
    elif members == 2:
        return "(3) 2"
    elif members == 3:
        return "(4) 3"
    elif members == 4:
        return "(5) 4"
    elif members == 5:
        return "(6) 5"
    elif members >= 6:
        return "(7) 6-higher"
```

# Missing values and Data treatment

We can noticed that some features or variables had some errors (Ex.: Age) and can hidden some information from us (Ex.: Name). To treat this We will use some data treatments to standardize the data and get some other features that can be useful for our analysis.

```python
# Name variable analysis
titanic_df['Name'].head()
```

The "Name" variable seems not be useful for us because it contain a probably unique value. But looking at our dataset it appears to be useful when deeply looked. We can see that every register contains a Title, like Mr., Mrs., Miss. and so one. Extracting this Title parameter it can be used to be compared to another features and why not with the Survived parameter.

```python
# Extract Nme Title
titanic_df["Title"] = titanic_df["Name"].str.extract('(\w*\.)', expand=True)
titanic_df.groupby(titanic_df["Title"]).size()
```

We know that null values can lower our analysis accuracy and even the confidence of our study. To solve this We can simply delete the rows with problem, but this will decrease the amount of data We have in our sample, or We can replace this null value with another value, it can be based on some other features, that what will be used by us here

```python
# Count the number of NaNs by column
print "- NaNs by column:"
print(pd.isnull(titanic_df).sum())
```

```python
# Replace missing Ages by Title mean
nan_ages=pd.isnull(titanic_df["Age"]).nonzero()

means_by_tile = titanic_df.groupby("Title").mean()['Age']

for i in nan_ages[0]:
    temp_title = titanic_df["Title"].ix[i]
    titanic_df.set_value(i, 'Age', int(means_by_tile[temp_title]))
```

```python
# To be easily treated, ages bellow 1 was replaced by 1   
titanic_df['Age'] = titanic_df['Age'].apply(normalize_age_lass_than_one)
```

We can ran again the describe function to "Age" field and now it will not return an error message.

```python
titanic_df['Age'].describe()
```

# Outlier treatment

To treat outliers We will only remove the rows with outliers. Some columns that can hold outliers: Age, SibSp, Parch, Fare.

```python
titanic_without_outliers = pd.DataFrame()

# Remove Age outliers
titanic_without_outliers = titanic_df[((titanic_df.Age - titanic_df.Age.mean()) / titanic_df.Age.std()).abs() < 3]

# Remove SibSp outliers
titanic_without_outliers = titanic_df[((titanic_df.SibSp - titanic_df.SibSp.mean()) / titanic_df.SibSp.std()).abs() < 3]

# Remove Parch outliers
titanic_without_outliers = titanic_df[((titanic_df.Parch - titanic_df.Parch.mean()) / titanic_df.Parch.std()).abs() < 3]

# Remove Age outliers
titanic_without_outliers = titanic_df[((titanic_df.Parch - titanic_df.Parch.mean()) / titanic_df.Parch.std()).abs() < 3]
```

```python
# Describe Titanic Without Outlier DataFrame
titanic_without_outliers.describe()
```

```python
# Describe Titanic DataFrame
titanic_df.describe()
```

We have found 15 outliers (891-876). We can now use our DataFrame without outliers.

```python
titanic_df = titanic_without_outliers
```
